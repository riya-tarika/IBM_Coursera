```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\91863\\Downloads\\Topic_Survey_Assignment (1).csv",index_col=0)
```

```python
pd.DataFrame(df)
df=df.sort_values(by=["Very interested"],ascending=False)
df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
p=100*(df/2233)
p
```

```python
labels=100*(df.sum()/2233)
df.plot(kind='bar',color=['#5cb85c','#5bc0de','#d9534f'],figsize=(20,8),width=0.8)
plt.title('Percentage of the respondents interest in the different Data Science Areas',fontsize=16)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.spines['right'].set_visible(False)
plt.gca().axes.spines['top'].set_visible(False)
plt.gca().axes.spines['left'].set_visible(False)
plt.xticks(fontsize=14)
plt.legend(fontsize=14,loc='best')
for i in plt.gca().patches:
    plt.text(i.get_x()+.01, i.get_height()-3, \
            str(round((i.get_height()/2233)*100, 2))+'%', fontsize=22,
                color='black')
```
